```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data from a CSV file (replace 'data.csv' with your CSV file)
file_path = "C:/Users/SAARTH CHAHAL/Downloads/best-fit.csv"
df = pd.read_csv(file_path)

# Extract x and y values from the DataFrame
x = df['X']
y = df['Y']

# Calculate the slope and intercept of the line of best fit using NumPy
slope, intercept = np.polyfit(x, y, 1)

# Create a line of best fit equation
line = f'Line of Best Fit: y = {slope:.2f}x + {intercept:.2f}'

# Create a scatter plot of the data points
plt.scatter(x, y, label='Data Points')

# Plot the line of best fit
plt.plot(x, slope * x + intercept, color='red', label=line)

# Add labels and a legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Show the plot
plt.grid(True)
plt.title('Line of Best Fit')
plt.show()
```
